```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# generate a random dataset with 1000 data points in 1d sampled from 2 different normal distributions
np.random.seed(420)
data = np.random.randn(1000, 1)
data[:500] += 3
data[500:] -= 3

# plot the data
sns.distplot(data, kde=False, bins=50)
plt.xlabel('x')
plt.ylabel('count')
plt.plot()
```

```python
noise = np.random.normal(0, .5, 1000)

data_noise = data.flatten() + noise.flatten()

sns.distplot(noise, kde=False, bins=50)
plt.xlabel('x')
plt.ylabel('count')
plt.show()

sns.distplot(data_noise, kde=False, bins=50)
plt.xlabel('x')
plt.ylabel('count')
plt.show()
```
