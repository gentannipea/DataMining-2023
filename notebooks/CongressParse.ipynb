```python
import pandas as pd
import numpy as np
import re
import json

CONGRESS = '../dataset/congress/legislators-historical.json'

with open(CONGRESS) as f:
    data = json.load(f)

terms = [leg['terms'] for leg in data]
terms = [item for sublist in terms for item in sublist]
```

```python
df = pd.DataFrame(terms)
df = df[df['type'] == 'sen']
df = df[df['party'].isna() == False]

useful = pd.DataFrame()
useful['party'] = df['party'].apply(lambda x: x.upper())
useful['state'] = df['state']
useful['district'] = df['district']
useful['year'] = df['start'].apply(lambda x: int(x.split('-')[0]))
useful
```
